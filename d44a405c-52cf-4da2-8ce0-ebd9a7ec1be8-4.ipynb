#### lr + tfidf

```python
%%time
#–æ–±—É—á–µ–Ω–∏–µ
regression_tfidf = LogisticRegression(fit_intercept=True, 
                                class_weight='balanced', 
                                random_state=42,
                                solver='liblinear'
                               )
#—Å–ª–æ–≤–∞—Ä—å
regression_parametrs = {'C': [0.1, 1, 10]}

#RandomizedSearchCV —Å –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏–µ–π
regression_grid = RandomizedSearchCV(regression_tfidf, regression_parametrs, scoring='f1', cv=3)
regression_grid.fit(features_train, y_train)

regression_tfidf.fit(features_train, y_train)
regression_cv_score = cross_val_score(regression_tfidf,features_train, y_train,scoring='f1',cv=3).mean()
print('–°—Ä–µ–¥–Ω—è—è –º–µ—Ç—Ä–∏–∫–∞ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –Ω–∞ –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏–∏:', regression_cv_score)
```

```python
%%time

regression_params_tfidf = regression_grid.best_params_
regression_score_tfidf = regression_grid.score(features_train, y_train)
print(regression_params_tfidf)
print(regression_score_tfidf)
```

#### rf + tfidf

```python
%%time
# –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è
forest = RandomForestClassifier(class_weight='balanced', n_jobs=-1 ,random_state=12345 )
#—Å–ª–æ–≤–∞—Ä—å 
forest_parametrs = { 'n_estimators': range(70, 75, 5),
                     'max_depth': range(20, 24, 4),
                     'min_samples_leaf': range(7,9),
                     'min_samples_split': range(4,6,2)}

# RandomizedSearchCV —Å –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏–µ–π
forest_grid = RandomizedSearchCV(forest, forest_parametrs, scoring='f1', cv=3)
forest_grid.fit(features_train, y_train)
```

```python
%%time
forest_params_tfidf = forest_grid.best_params_
forest_score_tfidf = forest_grid.score(features_train, y_train)
print(forest_params_tfidf )
print(forest_score_tfidf )
```

#### dt + tfidf

```python
%%time
# –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è
dt = DecisionTreeClassifier(class_weight='balanced',random_state=12345 )
#  —Å–ª–æ–≤–∞—Ä—å 
dt_parametrs = { 
                     'max_depth': range(20, 24, 4),
                     'min_samples_leaf': range(7,9),
                     'min_samples_split': range(4,6,2)}

# RandomizedSearchCV —Å –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏–µ–π
dt_grid = RandomizedSearchCV(dt, dt_parametrs, scoring='f1', cv=3)
dt_grid.fit(features_train, y_train)
```

```python
%%time
dt_params_tfidf = dt_grid.best_params_
dt_score_tfidf = dt_grid.score(features_train, y_train)
print(dt_params_tfidf)
print(dt_score_tfidf)
```
