```python
sample_1 = new.loc[new['platform'] == 'XOne', 'user_score'].dropna()
sample_2 = new.loc[new['platform'] == 'PC', 'user_score'].dropna()
results = st.ttest_ind(
    sample_1,
    sample_2,
    equal_var=False,
    alternative='two-sided'
)

print(results.pvalue)
if results.pvalue < 0.05:
    print('–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')
else:
    print('–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –º–∏–Ω—É—Ç—É')
```

–∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è H1: –°—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –∂–∞–Ω—Ä–æ–≤ Action (–∞–Ω–≥–ª. ¬´–¥–µ–π—Å—Ç–≤–∏–µ¬ª, —ç–∫—à–µ–Ω-–∏–≥—Ä—ã) –∏ Sports (–∞–Ω–≥–ª. ¬´—Å–ø–æ—Ä—Ç–∏–≤–Ω—ã–µ —Å–æ—Ä–µ–≤–Ω–æ–≤–∞–Ω–∏—è¬ª) –Ω–µ —Ä–∞–≤–Ω—ã

```python
sample_1 =  new.loc[new['genre'] == 'Action', 'user_score'].dropna()
sample_2 = new.loc[new['genre'] == 'Sports', 'user_score'].dropna()
if variance(sample_1) == variance(sample_2):
    print('–î–∏—Å–ø–µ—Ä—Å–∏–∏ —Ä–∞–≤–Ω—ã, –∏—Å–ø–æ–ª—å–∑—É–µ–º equal_var=True')
else:
     print('–î–∏—Å–ø–µ—Ä—Å–∏–∏ —Ä–∞–∑–ª–∏—á–Ω—ã, –∏—Å–ø–æ–ª—å–∑—É–µ–º equal_var=False')
```
